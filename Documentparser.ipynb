```python
from sklearn.datasets import load_files
```

```python
raw_data = load_files("/home/tathagat/machine-learning/ML/document-analytics-master")
```

```python
import sklearn
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline
import numpy as np
from sklearn import datasets
from pprint import pprint
count_vect = CountVectorizer()
X_train_counts = count_vect.fit_transform(raw_data.data)

# Fit the estimator and transform the vector to tf-idf
from sklearn.ensemble import GradientBoostingClassifier

tf_transformer = TfidfTransformer(use_idf=False).fit(X_train_counts)
X_train_tf = tf_transformer.transform(X_train_counts)
X_train_tf.shape
from sklearn.model_selection import train_test_split


tfidf_transformer = TfidfTransformer()
X_train_tfidf = tfidf_transformer.fit_transform(X_train_counts)
print X_train_tfidf.shape

train_X, test_X, train_y, test_y = train_test_split(X_train_tfidf, 
                               raw_data.target,
                                )
clf = GradientBoostingClassifier(random_state=42).fit(train_X, train_y)

print clf.score(test_X, test_y)
```
